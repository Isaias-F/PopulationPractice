```python
# Librer√≠as
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Usando pandas, lee el archivo csv
df = pd.read_csv('ve.csv')
# Elimina las columnas
df = df.drop(['lat', 'lng', 'iso2', 'capital', 'population_proper', 'country'], axis=1)

# Busca dentro de la columna el row
zulia = df.loc[(df['admin_name'] == 'Zulia')]
get_rows = zulia.head(5)  # Muestra la cantidad de rows
```

```python
x = get_rows['city']  # Etiquetas
h = get_rows['population'].astype(int)  #astype elimina los numeros decimales
```

```python
explode = (0.1, 0, 0, 0, 0)  # Separa la parte principal de la gr√°fica
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'cyan']

def func(pct, allvalues):  # Funcion que retorna el porcentaje y totoales
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n{:d}".format(pct, absolute)

plt.pie(h, labels=x, explode=explode, colors=colors, autopct=lambda pct: func(pct, h), startangle=60, pctdistance=1.25, labeldistance=1.45, shadow=True)  # Crea la gr√°foca
plt.title('Poblaci√≥n en Zulia')
plt.legend(title='Ciudades')
plt.axis('equal')
plt.show()
get_rows
```
